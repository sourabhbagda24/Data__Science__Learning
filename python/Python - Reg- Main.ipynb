# Python Identity Operators 
 Identity operators are used to compare the memory location of two objects, not their values.

**is Operator**

- Returns True if both variables point to the same object in memory.

- Returns False if they are different objects, even if their values are the same.

**is not Operator**

- Returns True if both variables do not point to the same object.

- Returns False if they do point to the same object.

```python
#Python Identity Operators

a = [1, 2, 3]
b = a          # both point to the same memory location
c = [1, 2, 3]  # new list, different memory location

print(a is b)   # ‚úÖ True (same memory address)
print(a is c)   # ‚ùå False (different memory addresses)
print(a == c)   # ‚úÖ True (values are same)

```

#Generator in Python: 
- Python generator is like an ‚Äúon-demand number machine‚Äù ‚Äî it only creates the next value when you ask for it, so memory consumption stays very low until a value is actually requested.‚Äã‚Äã

- In contrast, a normal Python list puts all values in memory at once; if your range is huge, the whole list is stored, using a lot of RAM.‚Äã‚Äã

#Normal Function vs. Generator
- In a normal function, you would:

- Use a loop from 1 to 100 (or even 1 crore), append every number to a list, and finally return the full list.

- If the number of items is very large, this ends up using a huge amount of memory since every item stays in RAM.‚Äã‚Äã

# With a generator:

- The function ‚Äúpauses‚Äù after each yield and creates the next number only when you specifically ask for it (using next() or a loop).

- Generator doesn‚Äôt keep every number in memory ‚Äî only the current value exists, so large or infinite sequences don‚Äôt slow down your system.‚Äã‚Äã

# Main Benefits of Generators
- Only one value in memory at a time (great for extremely large sequences).

- Saves memory, especially when handling big data or reading files line by line.

- yield keyword is used to create a generator function, which remembers its position and only continues where it left off.

```python
def make_list():
    nums = []
    for i in range(1, 10):
        nums.append(i)
    return nums
y = make_list()
print(y)        # Big memory usage!
```

```python
def gen_numbers():
    i = 1
    while i <= 1000000:
        yield i
        i += 1      # Only one value at a time!
x = gen_numbers()
print(next(x)) 
print(next(x))         
```

```python
def func():
    print("this is the first line")
    yield 1
    print("secodn")
    yield 2
    print("three")
    yield 3

f = func()
print(next(f))
print(next(f))
```

```python
# to generate fibonaci sereis through generator
def fibo(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a, b = b, a + b
x = fibo(7)
print(next(x)) 
print(next(x)) 
print(next(x)) 
print(next(x))
print(next(x))
print(next(x))
```

```python
def gen():
    x = 10
    print("Start generator")
    y = yield x       
    print("Received from send():", y)
    z = yield x + y
    print("Received second send():", z)

g = gen()

print(next(g))        
print(g.send(5))      
print(g.send(20))     
```

```python
def sample():
    print("Generator started")
    yield 1
    yield 2
    print("End")

g = sample()

print(next(g))  
g.close()       

print(next(g))   
```

```python
def test():
    try:
        yield 10
        yield 20
    except ValueError:
        print("ValueError handled in generator")
    yield 30

g = test()
print(next(g))           
g.throw(ValueError)     
print(next(g))           
```

# Normal Function (returns a value and ends)

A normal function:

- ‚úî Runs completely in one go
- ‚úî Returns one value
- ‚úî Variables and state are destroyed after returning
- ‚úî Memory usage high if you return a large list

#Generator Function (returns values one by one using yield)

A generator:

- ‚úî Doesn't run completely at once
- ‚úî Returns one value at a time
- ‚úî Maintains internal state
- ‚úî Uses less memory
- ‚úî Can be paused and resumed
- ‚úî Faster for large data

# ‚úÖ Iterable 

- Iterable means:
An object that can be used in a loop because it contains multiple elements.

Simple Explanation:

It is like a container that holds items.

You can loop through it using a for loop.

It has an __iter__() method.

But it does NOT have a __next__() method.

Examples:

list ‚Üí [1, 2, 3]

string ‚Üí "abc"

tuple ‚Üí (4, 5, 6)

One-line definition:

Iterable is an object that contains items and can be looped over.

# ‚úÖ Iterator (Easy English)

Iterator means:
An object that gives one item at a time when you call next().

Simple Explanation:

It is like a pointer or bookmark that remembers where you are.

It has both __iter__() and __next__() methods.

Each time you call next(), it returns the next element.

When elements finish, it raises StopIteration.
