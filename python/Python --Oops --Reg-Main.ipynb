```python
# Types of constructors in Python:
# 1. Default constructor: No arguments.
# 2. Parameterized constructor: Takes arguments.

class DefaultConstructorExample:
    def __init__(self):
        print("Default constructor called")

class ParameterizedConstructorExample:
    def __init__(self, value):
        self.color = value
        print("Parameterized constructor called with value:", value)

dc = DefaultConstructorExample()
pc = ParameterizedConstructorExample(10)

# Inheritance in Python:
# Inheritance allows a class (child) to inherit attributes and methods from another class (parent).

class Parent:
    def __init__(self):
        self.parent_attr = "I am parent"
    def parent_method(self):
        print("Parent method called")

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.child_attr = "I am child"
    def child_method(self):
        print("Child method called")

c = Child()
print(c.parent_attr)
c.parent_method()
print(c.child_attr)
c.child_method()   
```

# Encapsulation in Python

# Definition:
# Encapsulation is the concept of bundling data (attributes) and methods (functions) that operate on that data into a single unit, usually a class.
# It acts as a "shield" to protect the internal workings of a class from outside interference.

# Data Hiding:
# Encapsulation is also known as data hiding because it restricts direct access to certain data and methods within a class.

# Achieving Encapsulation:
# In Python, encapsulation is achieved using access specifiers:
# - Public: Accessible from anywhere.
# - Protected: Prefixed with a single underscore (_), intended for internal use.
# - Private: Prefixed with double underscores (__), not accessible directly outside the class.

# Accessing Private Data:
# - Direct access is restricted.
# - Can be accessed via public methods (getters and setters).
# - Name mangling (__attribute becomes _ClassName__attribute).

# Getters and Setters:
# - Getter methods retrieve the value of a private attribute.
# - Setter methods modify the value, often with validation.

# Advantages:
# - Security and data hiding
# - Improved code readability
# - Better maintainability

# Abstraction in OOP - Notes

- **Definition:**  
  Abstraction is the process of showing only the essential features of an object while hiding its complex internal implementation details from the user.

- **Purpose:**  
  - Simplifies complex systems by exposing only relevant functionalities.
  - Hides unnecessary details, making interaction easier.

- **Examples:**  
  - Searching for a YouTube video: Users interact with a simple search bar, not the underlying algorithms.
  - Human body: We perform actions (e.g., walking) without knowing the internal biological processes.

- **Generalization:**  
  - Abstraction allows general actions (e.g., a `Person` class with an `eat` method), while specific implementations can vary.

- **Design Level:**  
  - Abstraction is a thought process at the design stage, deciding what to show and what to hide.

- **Relation to Encapsulation:**  
  - Actual hiding of details is achieved through encapsulation, which restricts access to internal data and methods.

- **Key Points:**  
  - Focuses on "what" an object does, not "how" it does it.
  - Improves code modularity and maintainability.

# Access Modifiers in Python - Notes


there is no private adn protected method in py but we can act them as 
- **Abstraction and Security:**  

  Access modifiers are linked to abstraction and data hiding, which are crucial for security in programming.

- **Types of Access Modifiers:**  
  Three main types: public, protected, and private, common in many programming languages.

- **Public Access:**  
  By default, all attributes and methods in Python are public. They can be accessed and modified from anywhere, both inside and outside the class.

- **Protected Access:**  
  A single underscore prefix (e.g., `_roll_number`) denotes a protected attribute or method. Python doesn't strictly enforce this; it's a convention for good programming practice. Protected members can still be accessed and modified from outside the class or by derived classes.

- **Private Access:**  
  A double underscore prefix (e.g., `__age`) denotes a private attribute or method. These are intended to be accessible only within their own class and not by derived classes or from outside.

- **Name Mangling:**  
  Python uses name mangling for private attributes (e.g., `__age` becomes `_ClassName__age`). This makes private attributes technically accessible if one knows the mangled name, but it's a convention to avoid misuse.

- **Key Point:**  
  Python's access specifiers are about naming conventions and programmer responsibility, not strict enforcement, unlike languages like Java or C++.

```python
class student:
    def __init__(self, name, roll_no, age):
        self.name = name             
        self._roll_no = roll_no       
        self.__age = age              

    def __display(self):            
        print(f'Hi, my name is {self.name}, age is {self.__age}, and roll no is {self._roll_no}')

    def displayPrivateDisplay(self):  
        self.__display()

    def show_public(self):
        print(f'Public: {self.name}')

    def show_protected(self):
        print(f'Protected: {self._roll_no}')

    def show_private(self):
        print(f'Private: {self.__age}')

class branch(student):
    def show(self):
        print(f'Accessing protected in child: {self._roll_no}')
     
        print(f'Accessing private in child: {self._student__age}')

s = student("Alice", 101, 20)
b = branch("Bob", 102, 21)


s.show_public()
# Protected
s.show_protected()

s.show_private()

s.displayPrivateDisplay()

print(f'Name mangling: {s._student__age}')


b.show()
```

```python
import random

class bankaccount(ABC):   
    def __init__(self, name, bank_balance):
        self.accNo = random.randint(100, 200)
        self.name = name
        self.__bank_balance = bank_balance

    def accountBalance(self):  # getter method which returns private variable
        return self.__bank_balance

    def balance_update(self): # setter is used to modify the values
        self.__bank_balance -= 1
        return self.__bank_balance

    def deposit_amount(self, amount):
        if amount > 0:
            self.__bank_balance += amount
        else:
            print("we can not add money in -ve")  

    def withdraw(self, amount):
        if self.__bank_balance >= amount and amount > 0:
            self.__bank_balance -= amount   
        else:
            print("amount can not be withdrawn")    

b1 = bankaccount("yash", 100)
b2 = bankaccount("hars", 1000)
b1.deposit_amount(120)
b2.deposit_amount(30)
display(b2.accountBalance())
display(b1.accountBalance())
b1.withdraw(100)
b1.balance_update()
print(b1.accountBalance())
```

```python
class Rs_dollar:
    def __init__(self, amount):
        self.amount = amount

    def __repr__(self):
        return f'this is object of a rs to dollar class ${self.amount}'

    @classmethod
    def conversion(cls, x, y):
        return cls(x / y)

class Euro(Rs_dollar):
    def __init__(self, amount):
        self.amount = amount

    def __repr__(self):
        return f'this is object of a rs to euro class euro {self.amount}'

display(Rs_dollar.conversion(200, 85))
display(Euro.conversion(200, 90))
```
