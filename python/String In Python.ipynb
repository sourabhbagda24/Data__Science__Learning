:menas we can (replace,uppcase,lowecase,remove white space,replace string,split string) perform opetion with the strings

```python
a = "hello"
b = "WORLD"
c= "hello world"
print(a.upper())
print(b.lower())
print(c.strip())
print(c.replace("world","governmet"))
print(c.split(","))
```

```python
txt = "we are the so-called \"Vikings\" from the north."
print(txt)
print(txt.capitalize())
print(txt.casefold())
print(txt.center(50))
print(txt.count("e"))
print(txt.encode())
print(txt.endswith("north"))
print(txt.expandtabs(2))
print(txt.find("north"))
print(txt.format())

print(txt.index("north"))
print(txt.translate({32: 42}))
print(txt.isalnum())
print(txt.isalpha())
print(txt.isdecimal())
print(txt.isdigit())
print(txt.isidentifier())
print(txt.islower())
print(txt.isnumeric())
print(txt.isprintable())
print(txt.isspace())
print(txt.istitle())
print(txt.isupper())
```
