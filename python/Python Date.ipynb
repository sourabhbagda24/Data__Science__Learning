```python
import datetime

now = datetime.datetime.now()

print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
print("Hour:", now.hour)
print("Minute:", now.minute)
print("Second:", now.second)
```

```python
import datetime

now = datetime.datetime.now()

print(now.strftime("%Y-%m-%d"))   # 2025-11-03
print(now.strftime("%d/%m/%Y"))   # 03/11/2025
print(now.strftime("%B %d, %Y"))  # November 03, 2025
print(now.strftime("%I:%M %p"))   # 12:30 AM
```

```python
import datetime

date_str = "03-11-2025 14:45:00"
date_obj = datetime.datetime.strptime(date_str, "%d-%m-%Y %H:%M:%S")

print(date_obj)
```

```python
import datetime

d1 = datetime.date(2025, 11, 3)
d2 = datetime.date(2025, 12, 25)

diff = d2 - d1
print(diff.days)  # 52 days
```

```python
from datetime import datetime, timedelta

now = datetime.now()
print("Now:", now)

future = now + timedelta(days=10)
past = now - timedelta(hours=5)

print("After 10 days:", future)
print("5 hours ago:", past)
```

```python
from datetime import datetime, timezone, timedelta

utc_time = datetime.now(timezone.utc)
print("UTC:", utc_time)

india_time = utc_time + timedelta(hours=5, minutes=30)
print("India Time:", india_time)
```

```python
import datetime

d1 = datetime.date(2025, 11, 3)
d2 = datetime.date(2025, 11, 5)

if d1 < d2:
    print("d1 is earlier")
else:
    print("d2 is earlier")
```

```python
import datetime

dob = datetime.date(2002, 6, 15)
today = datetime.date.today()

age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
print("Your Age:", age)
```
