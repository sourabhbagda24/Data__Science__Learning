```python
# Create a new column named total_score, which is the sum of math score,
# reading score, and writing score.
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
# Create a new column named average_score (total score divided by 3)
df['average_score'] = df['total_score'] / 3
```

```python
# Add a column named result. A student passes if their average_score is 60 or
# above; otherwise, they fail. (Hint: Use a conditional function or lambda expression).
# # 4. Find the top 5 students based on their total_score.
df['result'] = np.where(df['average_score'] >= 60, 'pass', 'fail')
```

```python
# Find the bottom 5 students based on their math score.
df.sort_values(by='math score', ascending=True).head(5)
```
