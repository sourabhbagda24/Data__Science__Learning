```python
# broadcasting rule ==> broadcasting in numpy is a powerful mechanim that allows arithmetic operations on array of different shape and size

# 1. Dimension Padding: If the two arrays have different numbers of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.
# 2. Dimension Compatibility: Two dimensions are considered compatible if they are either:
# --> they are equal in size
# --> one of them has a size of 1

# 3. Error Condition: If in any dimension the sizes disagree and neither is equal to 1, a ValueError exception is raised, and broadcasting cannot occur.
```

```python
mat1=np.arange(1,10).reshape(3,3)
mat2=np.arange(10,19).reshape(3,3)
print(mat1)
print(mat2)
print(mat1+mat2)
# according to the rule no. 1 of the broadcasting . the dimensions of both matrices have same dimensions.
```

```python
print(mat1*mat2)
```
